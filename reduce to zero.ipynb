```python
def number_of_steps(num):
    steps = 0
    
    while num > 0:
        if num % 2 == 0:
            num //= 2
        else:
            num -= 1
        steps += 1
    
    return steps


# Test cases
print(number_of_steps(14))  # Output: 6
print(number_of_steps(8))   # Output: 4
print(number_of_steps(123)) # Extra test
```
