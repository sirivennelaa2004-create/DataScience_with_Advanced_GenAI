```python
from typing import List

class Solution:
    def shuffle(self, nums: List[int], n: int) -> List[int]:
        result = []
        for i in range(n):
            result.append(nums[i])     # x_i
            result.append(nums[i + n]) # y_i
        return result
obj = Solution()
output = obj.shuffle([2,5,1,3,4,7],3)
print(output)
```
